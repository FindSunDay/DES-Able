```python
import urllib.request
import pandas as pd
import json
```

```python
dms = pd.read_csv('Dataset/1.1DMS_Providers_Export_2020_08_20_23_56_29.csv')
ess = pd.read_csv('Dataset/1.1ESS_Providers_Export_2020_08_20_23_57_16.csv')
```

```python
dms_vic = dms[dms['State'] == 'VIC']
ess_vic = ess[ess['State'] == 'VIC']
```

```python
dms_name = set(zip(dms_vic['Title'], dms_vic['Location']))
ess_name = set(zip(ess_vic['Title'], ess_vic['Location']))
```

```python
dms_vic.shape,ess_vic.shape
```

```python
dms_vic['Type'] = 'DMS'
ess_vic['Type'] = 'ESS'
des = pd.concat([dms_vic,ess_vic])
```

```python
des['Title'].replace({"WISE Employment":"WISE Employment Ltd", \
                      "Atwork Australia Pty Ltd":"atWork Australia Pty Ltd" }, inplace=True)
```

```python
des.shape
```

```python
des.drop(['SiteCode','ContractType','StreamA','StreamB','StreamC','ServiceType','Unnamed: 17'],axis=1,inplace=True)
```

```python
des = des.drop_duplicates(subset=['Title','Location','Address','Specialisation'], keep='first')
```

```python
# des.sort_values(by='Title')
```

```python
des['Specialisation'].unique()
```

```python
des.shape
```

```python
des.head()
```

```python
des['New_Phone'] = 'none'
des['API_Detail'] = ''
```

```python
des.reset_index(inplace=True)
```

```python
des.drop('index',axis=1,inplace=True)
```

```python
for index, row in des.iterrows():    

    api_place = 'AIzaSyD69eksEStdVqffwHzc2L_Y5btC5ePv_Ls'
    
    place = des.iloc[index,7].replace("–","").replace(" ","%20")
    place = place.replace(" ","%20")
    find_place_id = ('https://maps.googleapis.com/maps/api/place/findplacefromtext/json?input='\
                +place+\
                '&inputtype=textquery&fields=place_id,opening_hours,geometry&key='\
                +api_place)
    
    response = urllib.request.urlopen(find_place_id).read()
    place_detail = json.loads(response)
    
    des.loc[index,'API_Detail'] = [place_detail] 
    
    if 'candidates' in place_detail.keys() and len(place_detail['candidates']) >0:
        place_id = place_detail['candidates'][0]['place_id']
    
    find_phone = 'https://maps.googleapis.com/maps/api/place/details/json?place_id='\
                +place_id\
                +'&fields=rating,formatted_phone_number&key='\
                +api_place

    response2 = urllib.request.urlopen(find_phone).read()
    place_detail2 = json.loads(response2)
    
    if 'formatted_phone_number' in place_detail2['result'].keys():
        des.loc[index,'New_Phone'] = place_detail2['result']['formatted_phone_number']
    
```
